```python
import os
import json
import pandas as pd
from pathlib import Path
```

```python
for i in range(5):
    with open("data{}user_ids_and_session_ids_training_{}.json".format(os.sep, i), "r") as source:
        data = json.load(source)
    
    for user, sessions in data.items():
        with open("data{}sessions_30music_training_{}.json".format(os.sep, i), "r") as source:
            training_sessions = json.load(source)

        all_tags = {}
        for session_id in data[user]["session_subset"]:
            for tags in training_sessions[str(session_id)]["tags_idxs"]:
                for tag in tags:
                    if tag not in all_tags:
                        all_tags[tag] = []

        # tag: [0, 0, 0, 4, 5, 2, 0] shows frequency in each session
        for session_id in data[user]["session_subset"]:
            session_tags = {}
            for tags in training_sessions[str(session_id)]["tags_idxs"]:
                for tag in tags:
                    if tag in session_tags:
                        session_tags[tag] += 1
                    else:
                        session_tags[tag] = 1
            
            for tag in all_tags:
                if tag in session_tags:
                    all_tags[tag].append(session_tags[tag])
                else:
                    all_tags[tag].append(0)

        print(all_tags)
        input('')
    break
```
